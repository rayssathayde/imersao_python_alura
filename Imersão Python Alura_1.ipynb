```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel('/content/Imersão Python - Tabela de ações_Python.xlsx', sheet_name='Principal')
df_principal.head()
```

```python
df_total_acoes = pd.read_excel('/content/Imersão Python - Tabela de ações_Python.xlsx', sheet_name='Total_de_acoes')
df_total_acoes.head()
```

```python
df_ticker = pd.read_excel('/content/Imersão Python - Tabela de ações_Python.xlsx', sheet_name='Ticker')
df_ticker.head()
```

```python
df_chatgpt = pd.read_excel('/content/Imersão Python - Tabela de ações_Python.xlsx', sheet_name='ChatGPT')
df_chatgpt.head()
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'})
df_principal
```

```python
df_principal['variacao_pct'] = df_principal['var_dia_pct']/100
df_principal['valor_inicial'] = df_principal['valor_final']/(1 + df_principal['variacao_pct'])
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtd_de_acoes'})
df_principal
```

```python
df_principal['variacao'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['qtd_de_acoes']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal
```

```python
df_principal['qtd_de_acoes'] = df_principal['qtd_de_acoes'].astype(int)
df_principal
```

```python
df_principal['resultado'] = df_principal['variacao'].apply(lambda x: "Subiu" if x > 0 else ("Desceu" if x < 0 else "Estável"))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Nome':'nome_empresa', 'Idade em Anos':'idade_empresa'})
df_principal
```

```python
df_principal = df_principal.rename(columns={'Ativo':'ativo', 'Data':'data', 'Segmento':'segmento'})
df_principal
```

```python
df_principal['cat_idade'] = df_principal['idade_empresa'].apply(lambda x: "Mais de 100 anos" if x > 100 else ("Menos de 50 anos" if x < 50 else "Entre 50 e 100 anos"))
df_principal
```

```python
df_principal = df_principal.drop_duplicates(keep='first')
df_principal.shape
```

```python
# Calculando os valores desejados
maior = df_principal['variacao'].max()
menor = df_principal['variacao'].min()
media = df_principal['variacao'].mean()
media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['variacao'].mean()
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['variacao'].mean()


# Imprimindo os resultados
print("Maior\t", 'R$ {:,.2f}'.format(maior))
print("Menor\t", 'R$ {:,.2f}'.format(menor))
print("Média\t", 'R$ {:,.2f}'.format(media))
print("Média de quem subiu\t", 'R$ {:,.2f}'.format(media_subiu))
print("Média de quem desceu\t", 'R$ {:,.2f}'.format(media_desceu))
```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu.head()
```

```python
df_analise_segmento = df_principal_subiu.groupby('segmento')['variacao'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('resultado')['variacao'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='resultado', y='variacao', text_auto='.2f', title='Variação (R$) por Resultado')
fig.show()
```

```python
fig1 = px.pie(df_analise_segmento, names='segmento', values='variacao', title='Variação (R$) por Segmento')
fig1.show()
```

```python
df_analise_idade = df_principal.groupby('cat_idade')['variacao'].sum().reset_index()
df_analise_idade
```

```python
fig2 = px.bar(df_analise_idade, x='cat_idade', y='variacao', text_auto='.2f', title='Variação (R$) por idade')
fig2.show()
```
